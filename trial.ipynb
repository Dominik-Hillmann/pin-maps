```python
import os
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader

from PIL import Image

from matplotlib.offsetbox import AnnotationBbox, OffsetImage
```

```python
width = 1500
height = 1500
dpi = 96
projection = ccrs.PlateCarree()

fig = plt.figure(figsize=(width / dpi, height / dpi), dpi=dpi)

file_path = os.path.join('data', 'de.shp')
shapes = list(shpreader.Reader(file_path).geometries())

ax = plt.axes(projection = projection)
#plt.title('SCHLAAAAAND')
#ax.add_geometries(
#    shapes,
#    projection,
#    edgecolor = 'none', 
#    facecolor = cfeature.COLORS['land'], 
#    # alpha=0.5
#)

ax.coastlines(resolution='10m')
#ax.stock_img()
ax.set_extent([5, 16, 47, 56], projection)
ax.add_feature(cfeature.NaturalEarthFeature(
    'physical', 
    'rivers_lake_centerlines', 
    '50m', # resolution, 
    edgecolor = cfeature.COLORS['water'], 
    facecolor='none'
))

print('Getting Netherlands')
netherlands_path = os.path.join('data', 'netherlands.shp')
netherlands_shape = list(shpreader.Reader(netherlands_path).geometries())
ax.add_geometries(
    netherlands_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)

print('Getting PL')
poland_path = os.path.join('data', 'poland.shp')
poland_shape = list(shpreader.Reader(poland_path).geometries())
ax.add_geometries(
    poland_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)


print('Getting DK')
dk_path = os.path.join('data', 'dk.shp')
dk_shape = list(shpreader.Reader(dk_path).geometries())
ax.add_geometries(
    dk_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)
print('Getting CH')
ch_path = os.path.join('data', 'ch.shp')
ch_shape = list(shpreader.Reader(ch_path).geometries())
ax.add_geometries(
    ch_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)
print('Getting F')
f_path = os.path.join('data', 'f.shp')
f_shape = list(shpreader.Reader(f_path).geometries())
ax.add_geometries(
    f_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)

print('Getting LUX')
lux_path = os.path.join('data', 'lux.shp')
lux_shape = list(shpreader.Reader(lux_path).geometries())
ax.add_geometries(
    lux_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)


print('Getting BE')
be_path = os.path.join('data', 'be.shp')
be_shape = list(shpreader.Reader(be_path).geometries())
ax.add_geometries(
    be_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)

print('Getting CZ')
cz_path = os.path.join('data', 'cz.shp')
cz_shape = list(shpreader.Reader(cz_path).geometries())
ax.add_geometries(
    cz_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)

print('Getting AU')
au_path = os.path.join('data', 'au.shp')
au_shape = list(shpreader.Reader(au_path).geometries())
ax.add_geometries(
    au_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)

print('Getting LIE')
li_path = os.path.join('data', 'li.shp')
li_shape = list(shpreader.Reader(li_path).geometries())
ax.add_geometries(
    li_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)
print('Getting SE')
se_path = os.path.join('data', 'se.shp')
se_shape = list(shpreader.Reader(se_path).geometries())
ax.add_geometries(
    se_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)
print('Getting IT')
it_path = os.path.join('data', 'it.shp')
it_shape = list(shpreader.Reader(it_path).geometries())
ax.add_geometries(
    it_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)

print('Getting Baltic sea')
baltic_sea_path = os.path.join('data', 'baltic.shp')
baltic_sea_shape = list(shpreader.Reader(baltic_sea_path).geometries())
ax.add_geometries(
    baltic_sea_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)

print('Getting North Sea')
north_sea_path = os.path.join('data', 'north.shp')
north_sea_shape = list(shpreader.Reader(north_sea_path).geometries())
ax.add_geometries(
    north_sea_shape,
    projection,
    edgecolor = 'white', 
    facecolor = 'white',
    zorder = 10
)


ax.set_aspect(1.35)


m = Image.open(os.path.join('data', 'wallpaper.jpg'))
imagebox = OffsetImage(m, zoom = 0.7)
imagebox.image.axes = ax
ab = AnnotationBbox(imagebox, (10.2, 52.1205), pad = 0, frameon = False)
ax.add_artist(ab)
#ax.plot(
#    11.5, 52, 
#    marker = 'v', 
#    markersize = 12, 
#    transform = projection
#)
ax.text(
    10.5, 51.2, 
    'Magdeburg', 
    transform = projection, 
    fontdict = {
        'color': 'red', 
        'family': 'serif', 
        'size': 20
    },
    zorder = 10
)

#img = plt.imread('emoji.png')
# print(img)
#ax.imshow(
#    img, 
#    #origin='upper', 
##    #transform = projection, 
#    extent = (11.5, 52, 11.5, 52.2),
#    zorder = 10
#)

# Control size via zoom
img = Image.open(os.path.join('data', 'emoji.png'))
imagebox = OffsetImage(img, zoom = 0.1)
imagebox.image.axes = ax
ab = AnnotationBbox(imagebox, (11.6276, 52.1205), pad=0, frameon=False)
ax.add_artist(ab)

m = Image.open(os.path.join('data', 'money-bag.png'))
imagebox = OffsetImage(m, zoom = 0.4)
imagebox.image.axes = ax
ab = AnnotationBbox(imagebox, (13.3923, 54.0865), pad=0, frameon=False)
ax.add_artist(ab)


plt.show()



# https://www.diva-gis.org/gdata
```
